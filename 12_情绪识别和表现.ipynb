# 1. 情绪分类法

 - 在HRI中，情绪通常使用环形模型进行分类，离散的情绪被放在模型上。 
 - 为此存在几个模型，有些还带有第三个 "强度 "维度。



# 2. 情绪描述

 - 机器人不体验情感，但有理由让机器人的行为描绘出情感。 
 - 情感描绘会有什么目的？
     - 内部状态显示 
     - 引起同理心
     - 改善互动 
     - 提高感知力

内部状态显示：
 - 许多机器人认知架构都有驱动机器人行为和对人类反应的状态。 
 - 让与机器人互动的人看到内部状态，可以更容易理解和预测机器人的行为。 
 - 让这些内部状态为人类所理解的一种方式是将它们描绘成情绪，例如，将错误状态描绘成悲伤，将感知的低信心描绘成困惑。



引起同理心

 - 同理心是人类对情感表现的自然反应。 
 - 有3种类型的同理心：
   - 认知性同理心（情感理解）。 
   - 情感性同理心（分享情感）。
   - 怜悯性移情（情感关注/帮助）。 
   - 引起同理心可以达到一些互动的目的：
     - 形成说服策略的一部分。 
     - 在需要时获得帮助。 
     - 为人类利益而产生情感影响（如helper's high，人际情感转移）。



改善互动
 - 由于人类习惯于他人表达情感，拥有一个会表达情感的机器人可以提高主观评价：智力、可塑性和拟人化。 
 - 情感可以使互动看起来更自然、更舒适、更有吸引力。


情感描述：显著的行为：
 - 描绘情绪的明确方式是使用明确的情绪表达--手势、面部表情或语言。 
 - 虽然这可以使观察者清楚地了解情绪，但它们会使互动看起来不那么自然/像人，并且需要中断其他交流行为。 
 - 面部表情是一个例外，因为它们明确地描绘了情绪而不一定会中断正常的互动行为。

情感描述：隐含的行为：
 - 来自人类情感理解的概念可以被用来修改机器人的行为以传达情感。一个已经被应用于HRI的模型是SIRE：速度、强度、规律、意图。 
 - 另一个可以使用的隐含线索是语言的音调。

# 3. 机器人的情绪识别

 - 为了做出适当的反应，机器人可能需要识别情感。 
 - 情感识别会有什么作用呢？ 
   - 根据人类的情绪状态修改行为。 
   - 人类的行为预测。 
   - 显示同理心。

识别：行为调整

 - 机器人的行为可以根据交互伙伴的情绪状态进行修改：
   - 为沮丧或困惑的用户修改指令的传递。 
   - 检测对帮助的需求。 
   - 改变对话结构/传递以保持参与。

识别：行为预测
 - 在许多情况下，人类行为是由我们的情绪状态驱动的。情绪识别可以提高机器人预测人类行为的能力，特别是在社会互动中。 
 - 围绕辅助机器人的参与和脱离的决定应该受到用户情绪的影响。

识别：显现同理心

 - 为了描绘同理心（如前所述），机器人必须理解其伙伴的情绪状态。 
 - 哪种形式的同理心应该被描绘出来是一个开放的研究问题。检测到的情绪和互动的背景都可能起到一定的作用。最近的一篇论文显示，情感上的共鸣而不是认知上的共鸣会导致互动的改善。

识别：面部表情
 - 自动情绪识别是一个活跃的研究领域--远远不是一个已经解决的问题！
 - 一个常见的方法是使用机器视觉处理面部，再加上机器学习。 
 - 一种方法是使用像OpenFace这样的软件来提取面部特征（例如使用FACS系统），然后使用机器学习将这些特征映射到人工注释的视频。
   - 面部动作编码系统（FACS）最初由Ekman和Friesen在1978年发表，并在随后的论文中进行了更新1。 它根据个别面部肌肉的可见效果将面部动作分类为动作单元（AU）。 它允许将任何面部表情分解为小的离散单元。 Openface能够识别大多数但不是所有的AU（那些容易从其跟踪点看到的AU）。它报告每个AU的存在和强度。

识别：其他的特征

 - 正如已经讨论过的，还有其他表达情感的方式，这些也被用于情感识别：
   - 处理情感语言的语言内容。
   - 机器学习将语音语调和身体语言特征映射到情感中。与面部表情一样，特征可以被自主提取，并被映射到手动编码的视频中。

识别：数据编码

 - 为了训练有监督的ML系统，需要给数据贴上不同情绪的标签。 
 - 通常这是由数据的人工编码完成的，这个过程中，视频（或音频）数据被多个注释者贴上标签，以便在每个片段的时间点上显示不同的情绪。 
 - 标签可以是连续的或离散的。

# 4. 伦理

 - 机器人的情感本质上具有欺骗性，因此提出了使用这些情感是否合适的伦理问题。 
 - 诱导人类产生共鸣，并通过人际情感转移等过程影响人类情感的过程具有未知的后果。
 - 情感和共鸣有可能作为一种强烈的说服工具，而这种说服的目标必须得到仔细考虑。

# 5. 总结
 - 情感分类法--情感的环形模型。 
 - 情感描写--用于显示内部状态，产生共鸣和改善互动；情感的显性和隐性描写。 
 - 情感识别--用于控制行为；可使用的数据包括身体语言、面部表情、语音语调；ML模型在标记的数据上训练。 
 - HRI中的情感伦理--情感描写具有欺骗性，情感操纵具有后果，共鸣具有说服力。
