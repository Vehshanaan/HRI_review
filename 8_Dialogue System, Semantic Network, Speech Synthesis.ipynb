# 2. 对话系统 / Dialogue System

 - 对话系统是能够通过口语互动帮助用户更有效地完成任务的智能代理。 
 - 对话系统正在被纳入各种设备中，如智能手机、智能电视、车载导航系统等。


什么是对话系统？

 - 口语对话系统是一个能够用语音与人交谈的计算机系统。 
 - 是聊天机器人？
   - 不是。它有两个在聊天机器人中不存在的基本组件：
     - 一个语音识别器和
     - 一个文本到语音模块。
 - 是command and control系统？
     - 不是。
       - 指挥和控制语音系统只能对请求作出反应，但不试图在一段时间内保持连续性。

定义

 - 计划与人类用户对话的系统 
 - 通常用于 
   - 自动电话服务 
   - 虚拟助手 
   - 多模态系统（如地图中的语音输入） 
   - 人与机器人互动 
 - 对话系统类型 
   - 基于有限状态的系统 
   - 基于框架的系统 
   - 基于代理的系统

相关技术

 - 基于有限状态的系统 
   - 系统引导用户完成对话步骤 
   - 使用固定的输入和输出语法 
   - 用户的自由度较小，对话步骤是固定的 
 - 基于框架的系统 
   - 系统填补框架的缺失部分（即系统需要的信息模板） 
   - 对话步骤不固定 
   - 系统需要更灵活的语法来解析用户的输入 
 - 基于代理的系统 
   - 系统和用户一起解决对话任务 
   - 系统使用逻辑推理来处理用户的回答

# 3. 语音合成的挑战：

1. 一词多义： 
    - "在贝斯的头上画了一个鲈鱼。" 
    - "看到画中的裂口，我流下了眼泪。" 
2. 数字：  
    - '30' vs '2019年3月30日' vs '1.30' 
    - '亨利八世' vs '第八章' 
1. 缩略语：宝马'VS'美国国家航空航天局'VS'先生' 
2. 语气：句子的语音旋律
