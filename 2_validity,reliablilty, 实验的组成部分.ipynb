# **有效性（validity）和可靠性（reliability）**



可复现但是没用，有用但是不可复现，既有用又可复现

 - 内部有效性：实验结果是只受实验**变量**的影响吗？
 - 外部有效性：实验结果能**泛化**到实验以外吗
 - 可靠性：如果我**重复**实验，能得到相同的结果吗？

有效性和可靠性的示例：盒子里的球
 - 实验的问题：哪个盒子里球多？
 - 内部有效性：
   - 不用数球，称盒子的重量就可以
   - 球的重量如果不均等呢？
   - 盒子的重量如果不等怎么办？
 - 外部有效性：
   - 实验结果是对世界上的所有盒子都为真（有效）吗？
 - 可靠性：
   - 一个个数球只在球不多的时候能用
   - 重复进行几次计数可以提高可靠性

对内部有效性有影响的东西：
 - order effects（顺序的影响）
   - 受试者表现越来越好，因为他们开始学会机器人的表现模式了
   - 受试者表现越来越差，因为他们累了
   - 实验任务的随机性很重要
 - selection effects（选择造成的影响）
   - 不要用已经存在的组中的受试者，如班级中的同学，他们的知识水平都是一个样
   - 随机分配受试者
 - experimenter bias（实验组织者的偏置）
   - 实验的组织者对其中的某些条件有偏好
   - 解决：
     - 用文字或视频进行引导
     - 对所有实验条件进行相同的说明
     - 双盲实验
   - 常见问题：
     - 问题或指导中有导向性
     - 问题中定义不确切
     - 错误的答案尺度（如不能回答不知道）

对外部有效性有影响的东西：
 - 受试者
   - 从目标组中选出受试者
   - 将已有的subgroup分开
 - 环境干扰
   - 在实验室环境中尽可能地浮现真实环境
   - 消除不属于实验的环境干扰
 - 介绍（Instructions）
   - 要模仿人与人之间的互动
   - 例：萝卜酒保应该和人类酒保的表现相仿
 - 任务
   - 用户的任务应当基于之前的上下文分析

对可靠性有影响的东西：
 - 不受控的变量
   - 用户在实验前的经验
   - 任务描述不准确
   - 测量误差
 - 尽可能消除不受控变量
   - 查询用户的先前经验
   - 精确地测量
 - 重复
   - 多来点受试者，多整几回
   - 用统计学手段评估数据

# **受控的实验**

1. 假设/Hypothesis
2. 自变量/Independent Variables
3. 因变量/Depedent Variables
4. 实验设计

## 假设：

 - 假设必须是可测试的（可量化且可测量）
 - 假设包含实验设计者的假设（assumption）

例：
 - 当机器人根据任务环境调整其语言时，它就会被视为一个更好的对话伙伴
 - 如果实施不当，自适应语音输出可能会导致用户的困惑。

## 自变量

 - 自变量就是随着实验条件变化而变化的量
 - 例如，两套机器人行为举止

例：
 - 机器人对受试者的两种回应方式


## 因变量

 - 测量因变量来测试假设
 - 客观变量：时间，counting events
 - 主观变量：问卷，满意度调查

例：
 - 分为四个大类的问卷
   - 机器人的智能度
   - 任务的难易度
   - 用户的感受
   - 会话的质量
 - 目标：
   - 统计实验视频中的事件
   - 日志
   - 用户和机器人的回合（turn），对话长度，正确构建对象的数目

## 实验设计



平衡（counterbalancing）技巧

 - 用在受试者身上的一种技巧
 - 用**latin square**给受试者分配任务
 - 消除顺序影响（order effect）
 - 很适合受试者数量不多的情况
 - 保证实验的每个条件在每个位置出现的频率相同
