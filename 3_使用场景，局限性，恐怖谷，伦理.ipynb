# 2. 机器人的局限性

 - 智能优先/缺乏常识：适应能力差
 - 传感器有限：传昂起一般只能感知到有限的东西，视觉系统也只能追踪它们跟踪的feature
 - 一般来说只专精于一个领域
 - 可人机交互的系统为了保证安全性，出力一般很低

局限性的后果：
 - 主观评价会受影响
 - 行为不完美时，观察者会自动填补空白
 - 可以忽略一些不被期望的功能：如，情感感知（portrayal）
 - 机器人的外观和行为设计可以用来减少这种能力和要求不匹配带来的mismatch感——不要试着设计得太好
 - 绿野仙踪法可以利用对能力的过度估计（over estimation of capabillities）



# 3. 说了一下恐怖谷

恐怖谷的原因：
 - 对人类形态的violaiton：如果一个玩意儿完全不像人，那它的人类特征会是显而易见的。但是如果一个玩意儿看着非常像人类，那就触发了我们对其他人类的模型
 - 冲突的认知线索：个体人知道类别隶属性的冲突线索时，就会产生冲突的认知表征
 - 死亡的显著性：机器人令我们的死亡更加显著，引发了对死亡的天生恐惧

对HRI的影响？

 - 产生负面的主观评估
 - 选择机器人平台和设计其行为的时候记着点，别太像人
 - 有些恐怖谷元素是可以测试的
 - 虽然有很多恐怖谷存在的证据，但是还没法确定它存在
  

# 4. 机器人部署的伦理学

几个关键因素：
 - Deception/欺骗   
 - Atachment/依附
 - Persuasion/说服
 - Privacy/隐私
 - Liability/责任


欺骗：
 - 如果机器人的行为不能描述机器人的内部状态，那他就可能有欺骗性，且它的表达里可能是有获利动机
 - 欺骗是有问题的，因为欺骗会导致对机器人能力的错误信念
 - 人们可能不想被机器人欺骗
 - 有证据表明，一些被视为欺骗的行为对观察者来说是可以接受的

依附：
 - 情感依附是你从被易联众获得情感联系，我们对人，宠物，人际关系，地点，物品等都有依恋，可以与机器人形成依恋关系
 - 由于机器人可能会损坏或被带走，因而存在伦理问题
 - 由于机器人的人工性质，依恋也会损害判断能力，并产生心理影响

说服：
 - 是社交辅助机器人的基础需求
 - 为说服人们而设计的系统应当仔细考虑
 - 社交辅助机器人的设计要考虑到这部分，因为它不应该说服人类进行有害的活动

隐私：
 - 随着机器人装备了大量的传感器，保护隐私就成了一个大麻烦
 - 当人类观察数据的时候，比如远程呈现机器人，就会加剧这个问题
 - 处理机器人数据需要健全的隐私政策

责任
 - 在涉及机器人的事故里，该怪谁？
 - 建议机器人配备伦理黑匣子，持续记录传感器和内部状态数据，来重现机器人为何以及如何造成事故
