GPT的训练步骤

1. 收集展示数据，训练一个受监督的政策
   - 从提示数据集中手机提示数据
   - 打标签者示范所有想得到的输出
   - 这些数据用于通过受监督的学习对GPT进行微调
2. 收集比较数据并训练奖励模型
   - 对提示和几个模型输出进行采样
   - 打标签者把这几个输出从好到差排个序
   - 这些数据用来训练我们的模型
3. 用PPO强化学习算法优化针对奖励模型的策略
   - 从数据集中采集一个新的提示
   - 从受监督政策中初始化一个PPO模型
   - 政策生成一个输出
   - 奖励模型计算此输出的奖励
   - 奖励被PPO用于更新其策略



# 6. NLP

**NLP金字塔**



从上至下依次为：语用学，语义学，句法，形态学

 - Morphology：基础级。处理单个的词。只关注词。检查词的拼写，等等。
 - Syntax：一般关注语法。检查句子的构成是否正确。
 - Semantics：关注句子的意思。仍然在句子的层级。这个句子是叙述还是问题？探测名字，实体等。
 - Pragmatics：包含以下层级的一切。他运作在文章层级。总结文章的一切。

Syntax等级的NLP：



1. POS tagging: Part of Speech Tagging
 - 检测一个词的词性（如形容词、名词、动词）
 - 有时被称为 "语法标签 "或 "词类消歧"
 - POS标签是大多数NLP系统的基本输入
 - 可以用基于规则和统计的系统实现
 - 现在主要是统计处理。 
  

2. Chunking
 - 识别句子的各部分 
 - 识别
    - 专有名词
    - 描述
    - 短语
    - 句子结尾 
  - 比全句解析更容易 
  - 有时被称为 "浅层解析"
  - 简单的Chunking可以用POS tagger和正则表达式实现

# 5. Parsing / 解析

 - 解析器：将输入转化为适合进一步处理的格式的程序 
 - 在语音的情况下，解析指的是语法分析 
 - 书面或口头语言的解析 
 - 将语言转移到逻辑表达中 
 - 寻找短语之间的上下文
 - 解析算法：
   - 自顶向下
   - 自底向上
 - 无上下文vs上下文敏感的语法
 - 语法的类型
   - 构成性语法 
   - 依赖性语法 
   - 统一性语法 
   - 类别性语法
