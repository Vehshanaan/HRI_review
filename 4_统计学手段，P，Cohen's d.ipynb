# 2. 统计学测试（statistical test）的四步
1. 数据可视化
2. 选择正确的统计学测试
3. 计算统计学显著度（significance）
4. 转译（interpretation）测试结果

数据可视化：
 - 目标：获得对数据的第一印象
   - 数据分布
   - 不同条件下数据的不同
   - 数据中的模式
 - 常用：带有whisker的bar图
   - 平均值用盒子表示
   - Whisker：误差条显示为与其中一个轴平行的线。这条线代表了该点的相应坐标的不确定性或变化。通常情况下，标准偏差被用来作为误差条。

选择统计学测试
 - 数据分布
   - 数据一般都遵循正态分布（高斯分布）
   - 很多统计学测试只在数据符合正态分布的时候才有效，这种叫参数测试（parameter test）
   - 当数据服从正态分布时：用参数测试
   - 当数据不服从正态分布时：用非参数测试
 - 正态分布
   - 直方图很好地描述了分布形式，如果你至少有一个中等规模的样本，你已经可以从直方图中确定是否满足正态分布的前提条件了。 
   - 除了对直方图进行目视检查外，还应该进行数学检查。为此，可以采用科尔莫戈罗夫-斯米尔诺夫检验和夏皮罗-威尔克检验。


计算显著性（**P value**）
 - 统计假设检验（statistical hypothesis test）对结果做一个假设，称为null hypothesis
 - 我们此处的null hypo是假设两组的均值没有差异
 - null hypo成立的概率用P表示
 - 如果某测试的P过低，则此假设可以抛弃。也就意味着两组之间存在统计学层面上的差异
 - 选择正确的测试：
   - 服从正态分布：
     - 2组：t-test
     - 2组以上：ANOVA
   - 不服从正态分布
     - 2组：Mann-Whitney 测试，Wilcoxon rank sum 测试
     - 2组及以上：Kruskall-Wallis H测试

结果转译（interpretation）
 - 要转译P，你得选一个显著性等级
 - 例如：10%(0.1), 5%(0.05), 2%(0.02)
 - 很多时候，**P<0.05**就代表有显著差异
 - 0.05的意思是，只有5%的概率，在被观测的分布之中，null hypo是真的（null hypo:两组之间没有差异）
 - 小心了，这是一个随意选择的显著性等级

效果大小/Effect Size
 - 效果大小是对实验效应大小的定量度量，效应值越大两变量之间的关系越强
 - Cohen‘s d是一个比较合适的用于比较两个均值的效应量度量。它可以和t-test以及ANOVA一起用
 - COhen’s d是两个均值之差除以数据的标准差
$$
 d = \frac{\bar{x}_1-\bar{x}_2}{s}
 $$
