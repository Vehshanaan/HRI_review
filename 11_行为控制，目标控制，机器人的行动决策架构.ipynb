# 2. Subsymbolic Control (Behavioural Control)

 - 核心思想：刺激触发行为，行为影响世界
 - 行为是读取传感器和控制执行器的小程序 
 - 每个行为做一件简单的事情 
 - 通常每个行为都可以访问所有的传感器

 - 问题：一次只想要一种行为 
   - 需要确定优先次序 
   - 行为可以覆盖或取代不太重要的行为
   - 每当优先级较高的行为变得活跃时，就将优先级较低的行为归入其中，从而实现行为的组合化




总结：

 - 关键词：行为，反应性的
 - 好处：允许增量开发，适应性强→在动态环境中有效，通常用于移动/反应型机器人
 - 坏处：面向目标的行为难以实现，调试困难
 - 风格：自底向上
 - 适应性：对于某些场景可以不做调整而适应
 - 扩展性：易于扩展，难以控制
 - 例子：subsumption architecture

# 3. Symbolic Architectures

- 符号/商议架构的基本范式 
- 将一个高层次的目标（"请给我做一杯我最喜欢的茶"）转化为原始的行动。



总结：
 - 关键词：商议式的/deliberative
 - 好处：
   - 非常有道理（high level reasoning）
   - 对互动有完全的控制
   - 更适合HRI的应用
 - 坏处：
   - 对于所有应用都必须重新规划架构
   - 很难扩展
 - 风格：自顶向下
 - 适应性：更改使用环境后，很多地方都要改
 - 扩展性：允许加入新的模块，但是老模块必须随着改
 - 例子：“Sense-Plan-Act”, JAMES Architecture
