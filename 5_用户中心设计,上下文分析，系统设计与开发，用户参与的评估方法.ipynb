# 2. 上下文分析/Context Analysis

分三块：
 - 用户：谁是系统的用户
 - 任务：用户想完成的是什么任务
 - 环境：用户界面是用在什么环境下


用户分析
 - 用户怎么描述他们自己的？
   - 用户怎么描述他们的职业？
   - 他们干活用的是什么工具？
 - 一个用户界面的不同用户之间有什么不同？
   - 个人差异
   - 偏好
   - 文化差异
   - 动机、目标、信念
 - 用户是怎么用界面的？
   - 用户具有什么样的专业水平？ 
   - 他们需要帮助功能吗？
   - 他们需要支持吗？
 - 分三步
   - 用户群组
     - 几乎没有只有一个用户群的情况 
     - 大多数时候，用户群不超过五个 
   - 对用户群体进行优先排序 
     - 思考对用户群体进行排序的标准（也许是排除）
     - 对两个（最多三个）群体进行优先设计
   - 为优先群体设计
   - 不要忘记，用户应该对设计过程有影响。


任务分析
 - 用户的目标是什么？
 - 用户现在用什么来实现他们的目标？ 
   - 他们用哪些工具来完成任务？ 
   - 如果新的用户界面不能正常工作，用户会回到旧的工具上
 - 用户为实现他们的目标正在执行哪些任务？
   -  现在执行的是哪些任务？
   -  执行任务的步骤是什么？
   -  执行任务需要多少时间？ 
   -  是否有导致任务执行中断的因素？
 - 任务不等同于系统功能
 - 用户考虑的是任务，而不是功能！ 
   - 任务："我必须改变史密斯先生的电话号码。" 
   - 功能："我必须在硬盘上找到史密斯先生的文件，把它加载到Excel中，......"
 - 常见的任务特征 
   - 任务步骤的顺序 
   - 步骤之间的依赖关系 
   - 重复的步骤 
 - 任务通常被明确定义，但没有考虑环境影响

环境分析
 - 物理环境
   - 用户有多少工作空间？ 
   - 是否有噪音或不断变化的照明条件？ 
   - 是否有足够的电力和互联网连接？ 
 - 社会环境 
   - 是否有很大的压力或时间压力？ 
   - 是否有足够的资源来帮助解决问题？ 
   - 人们如何一起工作？他们分享什么信息？ 
 - 文化环境 
   - 用户的文化背景有什么影响？ 
   - 专业或社会经济影响有什么影响？

上下文分析的方法
 - 在不与用户互动的情况下进行观察 → 例如，通过设置摄像头 
 - 有用户互动的观察 
   - 人种学访谈 
   - 情境调查 
   - 过程分析 
 - 如果不能进行观察，可以采用其他方法 
   - 用户日志 
   - 经验抽样 
   - （文化）探究 
   - （人工/artifact）演练 
 - 访谈 
   - 焦点小组 
   - 现场访谈 
 - 文献分析

# 4. 用户中心的评估

什么时候进行用户测试？
 - 理想情况下，用户测试在界面设计的所有阶段都是反复使用的 
 - 在设计过程之前 
   - 测试类似的界面 
   - 测试现有的界面（如果它们应该被替换） 
 - 在设计过程的开始 
   - 评估低保真度的原型 
 - 在设计过程中 
   - 测试不同的设计 
   - 测试设计的改变 
 - 在设计过程的结束 
   - 检查可用性目标是否已经达到 
   - 与其他界面的比较

用户评估的一般步骤：
1. 测试准备 
2. 编写测试计划 
3. 准备测试材料和环境 
4. 试点测试 
5. 参与者招募 
6. 参与者信息和同意 
7. 测试执行 
8. 测试分析和报告

用户中心的评估方法
 - 隐性方法（关注用户无法口头表达的内容）
   - 眼球跟踪/注意力跟踪
   - 用户跟踪
   - 任务和身体反应测量
   - 脑电图
 - 显性方法（用户反映他们的想法和感受）
   - 同时大声思考
   - 经验采样
   - 可用性测试
   - 用户研究
 - 基于专家的方法
   - 认知穿行
   - 启发式评价
